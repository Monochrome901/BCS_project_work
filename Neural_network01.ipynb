```python
#regresssion with neural network in tensorflow
import tensorflow as tf
print(tf.__version__)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#create features
X=np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0])

#create labels
Y=np.array([3.0,6.0,9.0,12.0,15.0,18.0,21.0,24.0])

#visualize it
plt.scatter(X,Y)
```

```python
Y==X+10
```

```python
input_shape=X.shape
output_shape=Y.shape
input_shape,output_shape
```

```python
#turn our numpy arrays to tensors
X=tf.cast(tf.constant(X),dtype=tf.float32)
Y=tf.cast(tf.constant(Y),dtype=tf.float32)
X,Y
```

```python
input_shape=X[0].shape
output_shape=Y[0].shape
input_shape, output_shape
```

```python
#steps in modelling with tensorflow
#set random seed
tf.random.set_seed(42)

#1. Create a model using sequential API
model=tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])
#2. Compile the model
model.compile(loss=tf.keras.losses.mae, #mae is short for mean absolute error
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"]) 
#3. Fit the model
model.fit(tf.expand_dims(X,axis=-1),Y,epochs=5)
```

```python
model.predict([17.0])
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train,Y_train,c="b",label="training data")
plt.scatter(X_test,Y_test,c="g",label="testing data")

plt.legend()
```

```python
#vizualizing our model's predictions
Y_pred=model.predict(X_test)
X_test,Y_pred
```

```python
#lets create a plotting function
def plot_predictions(train_data=X_train,
                    train_labels=Y_train,
                    test_data=X_test,
                    test_labels=Y_test,
                    predictions=Y_pred):
  #plots training data,test data and compares predictions to ground truth labels
  plt.figure(figsize=(10,7))
  plt.scatter(train_data,train_labels,c="b",label="training_data")
  plt.scatter(test_data,test_labels,c="g",label="testing_data")
  plt.scatter(test_data,predictions,c="r",label="predictions")
  
  plt.legend()
```

```python
plot_predictions()
```

```python
#regression evaluation
#mean absolute error(mae), great starter metric for any regression problem
#mean squared error(mse), when large errors are more significant than smaller errors
#Huber, combination of mse and mae
```

```python
def mae(Y_true,Y_pred):
  return tf.metrics.mean_absolute_error(Y_true=Y_true,Y_pred=tf.squeeze(Y_pred))

def mse(Y_true,Y_pred):
  return tf.metrics.mean_squared_error(Y_true=Y_true,Y_pred=tf.squeeze(Y_pred))
```
